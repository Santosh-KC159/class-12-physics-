### 3.1 Tone.Note,Harmonics and Overtones

When a body of certain length is vibrated, the sound wave of single frequency is produced.This sound wave single frequency is known as tone.The sound produced by single fork is pure tone.The tone of minimum frequency is called fundamental tone.

If the sound contains many frequencies,it is called note i.e. the combinations of many tones gives a note.The sound produced from a vibrating guitar is an example of note.


The sound waves in the frequency patterns vary with some regular sequence are known as harmonics.The harmonics incorporate fundamental tone and upper tones.The sound waves with frequency patterns above the fundamental tones are called overtones.Overtones are also called upper tones.

### 3.2 Waves in Pipes

#### Organ Pipe

A tube made of an material used to produced musical sound by setting air column in to vibrations is called an organ pipe.Flute,Whistle,Saxophone clarinet,etc. are organ pipes.It is a pipe in which lognitudinal stationary waves are formed when air is sent through it.Due to the formation of stationary waves,pipe emits sound of different fequencies.Organ pipe is of two types:
 


i.Closed organ pipe

ii.Open organ pipe 

##### Closed Organ Pipe

An organ pipe whose one end is closed and another end is open is called closed organ pipe or closed end pipe as shwon in the fig.For example,cap of a pen,bottle,key with hole etc.


(When air is blown from the open end,sound waves travel into the pipe so that the compression and rarefactions occurs.As the waves collides the close end, it reflects back and reflection phenomenon of sound waves holds.In this phenomenon incident wave and reflected wave superimpose and longitudinal wave is produced into the pipe.As the stationary wave is formed, node at closed end and antinodes at open end formed in to pipe that determines the frequency of sound.The audible sound is generated into the pipe only when the frequency of vibration produced by external source is equal to the natural frequency of vibration of air particles in to the pipe.)


###### Modes of vibration of closed organ pipe:

### Open Organ Pipe

An organ pipe whose both ends is open is called open organ pipe.For example,flute,clarinet etc.



When a sound source is held near the open end of the pipe, the air column is set into vibrations and a series of alternate compressions and rarefactions is generated within the pipe.These waves,as they reflect from the open end,come back to the vibrating system.When compressions reaches an open end,it is instantaneosly converted into rarefactions because there being nothing to prevent forward compression.At open end the density of air decreases and hence,rarefaction is formed which travels in the pipe.In this way,compression is reflected as rarefaction and similarly when a rarefaction reaches an open end,it is instantaneously converted into compression and is reflected back.Thus,reflected waves meet incident waves and give rise to stationary waves, and hence pipe produces sound of various frequencies.

In the case of open organ pipe,two open ends have to be antinode due to free vibration of air molecule there.

##### Modes of vibration of open organ pipe:

The different mode of the vibration in the closed organ pipe is discussed as below.

### End Correction

In general,it was assumed that the antinodes are formed exactly at the open end of the pipe.But it is observed that antinodes actually occurs at a little above or farther the end because the air molecules vibrate most freely at these locations slightly above the end.So,the vibrating length of air column is slightly greater than the length of pipe.Therefore correction factor has to be added to the length of pipe,which is know as end correction.

###### Thus, the distance between the end of the tube and the real positions of antinode outside it is called the end correction.it is denoted by c.

According to Lord Rayleigh, $c=0.6r =0.3 D$ for edgeless pipe,where $r$ is internal radius and $D$ is the internal diameter of pipe.This relation is called Rayleigh criteria.Thus,wider the pipe,greater the end correction.Also end correction depends on the wavelength $\lambda$ of the note and tends to vanish for very short wavelength.

$\therefore$ Effective wavelength of vibrating ait column =$l+0.6r$

Where $l$ is length of pipe and$r$ is its radius.

###### For closed pipe

$l$ is replaced by $l+c =l+0.3D$

###### For open pipe

$l$ is replaced by $(l+2c) =l+0.6 D$ , since in open pipe two end corrections are required.

#### Velocity of Transverse Wave along a stretched string

The formula for velocity of transverse wave along stretched string can be deduced by the method of dimension as follows

Let the velocity $v$ depends on the values of tension $(T)$, length $(l)$ and mass $(m)$ of the string along which waves travels.So,we can write,

$$v \propto T^{a},  v \propto l^{b} and v \propto m^{c} $$

Combining these relations,we get

$$v = kT^{a}l^{b}m^{c}    ............(i)$$           

Where a,b and c are numbers and k is a dimensionless constant whose experimental values is one.Writting dimension of each term,we get,

$$ \therefore  [LT^{-1}] = [M^{a+c} L^{a+b} T^{-2b}]$$

Equating the indices of $M,L,T$ on both sides,we get,

For M, $$a+c = 0$$,

For L, $$a+b =1$$

For T, $$-2a =-1$$

$$ \therefore  a= \frac{1}{2} , b= \frac{1}{2},  c=- \frac{1}{2} $$

putting these values in $(i)$,we get

$$v = k T^{\frac{1}{2}}l^{\frac{1}{2}}m^{-\frac{1}{2}}$$

$$ \therefore v = k \sqrt{\frac{Tl}{m}} = k \sqrt{\frac{T}{\frac{m}{l}}}$$

$$ \therefore v= k \sqrt {\frac{T}{\mu}}   ...........(ii)$$                   

Where $ \mu = \frac{m}{l} $ is mass per unit length of the string ot linear density of te string,and for k=1

But $v=f \lambda$.So from (ii),we get

$$f\lambda =\sqrt {T}{\mu}$$

$$ \therefore f = \frac{1}{\lambda} {\sqrt{T}{\mu}}   ..........$$      

Equation $(iii)$ gives frequency of Transverse wave in stretched string.



#### Modes of Vibration of Stretched String

If a string stretched between two points A and B is plucked,then transverse waves travel along the wire, and is reflected from fixed ends.The stationary waves are thus set up in the string and then the wire emits vibrations of different frequencies which are called modes of vibration as shown in fig.




###### First mode of vibration:

In this mode of vibration,two nodes and one antinode are formed in the string as shown in figure first.If $l$ be the length of string,$\lambda_{1}$ is the wavelength of vibration,$f_{1}$ is its vibration,then velocity of wave is

$$ v= f_{1} \lambda_{1}$$
$$ or, f_{1}=\frac{v}{\lambda_{1}} .......(i)$$

From $1^{st}$ fig.,it is clear that,

$$ l=\frac{\lambda_{1}}{2}$$

$$\therefore \lambda_{1}=2l  ........(ii)$$

from equations (i) and (ii),we get,

$$f_{1} =\frac{v}{2l}  .........(iii)$$

Again,

$$ v = \sqrt \frac{T}{\mu}   ..........(iv)$$

Now,equation $(i)$ can be written as,

$$ f_{1} = \frac{1}{2l} \sqrt{\frac{T}{\mu}} .........(v)$$

This is the fundamental frequency or frequency of 1st harmonics.

###### Second mode of vibration:

 In this mode of vibration the string vibration in two segments. Thus there are three nodes and two antinodes between of string as shown in figure second.If $\lambda_{2}$ be the wavelength and $f_{3}$ is the frequency of the wave in this mode of vibration,we have

$$f_{2} =\frac{v}{\lambda{2}} $$

$$ \therefore f_{2}=\frac{1}{\lambda_{2}} \sqrt{\frac{T}{\mu}}   ........(vi)$$

from $2^{nd}$ fig.,we have,

$$l=\frac{\lambda_{2}}{2}+\frac{\lambda_{2}}{2} =\lambda_{2}     .........(vii)$$

So,from $(vi)$ and $(vii)$,we can write,

$$f_{2}=\frac{1}{l} \sqrt{\frac{T}{\mu}}=2*\frac{1}{2l}\sqrt{\frac{T}{\mu}}$$

$$ \therefore f_{2}=2f_{1}    ...........(viii)$$

The frequency of vibration produced in this mode is twice the fundamental frequency and called octave.The vibration produced in this mode is called second harmonic or first overtone.

###### Third mode of vibration:

In this mode of vibration the string vibration in three segments. Thus there are four nodes and three antinodes between of string as shown in figure third.If $\lambda_{3}$ be the wavelength and $f_{3}$ is the frequency of the wave in this mode of vibration,we have

$$f_{3} =\frac{v}{\lambda{3}} $$

$$ \therefore f_{3}=\frac{1}{\lambda_{3}} \sqrt{\frac{T}{\mu}}   ........(ix)$$

from $3^{rd}$ fig.,we have,

$$l=\frac{\lambda_{3}}{2}+\frac{\lambda_{3}}{2}+\frac{\lambda_{3}}{2} =\frac{3\lambda_{3}}{2}$$

$$ or,\lambda_{3}=\frac{2l}{3}   ..........(x)$$

So,from $(ix)$ and $(x)$,we can write,

$$f_{3}=\frac{3}{2l} \sqrt{\frac{T}{\mu}}=3*\frac{1}{2l}\sqrt{\frac{T}{\mu}}$$

$$ \therefore f_{3}=3f_{1}    ...........(xi)$$

The frequency of vibration produced in this mode is thrice the fundamental frequency and so this vibration is called third harmonic or second overtone,In this way, for the $4^{th}$, $5^{th}$ …… modes of vibration in the stretched string the frequency for the $4^{th}$, $5{th}$….. Harmonics are emitted.

The general equation of frequency of wave in stretched string is given by,

$$ f_{n}=\frac{n}{2l} \sqrt{\frac{T}{\mu}}$$

$$ or, f_{n}=nf_{1} $$

where $n=1,2,3,......$ for $1^{st}$ ,$2^{nd}$ ,$3^{rd}$ ......harmonics respectively.



#### Laws of Transverse Vibration of String

The velocity of a transverse wave vibrating in a stretched string is given by

$$ v = \sqrt \frac{T}{\mu}       ..............(i)$$

Where T be the tension of string and μ is the mass per unit length. Since the frequency, $f=\frac {v}{2L}$ in fundamental mode, then from above expression we have

$$ f=\frac{1}{2l}\sqrt{T}{μ}      …………………(ii)$$

From equation (ii) we have followed the relation and this relation brings laws of transverse vibration of stretched string.

###### a. law of length (first law):

The fundamental frequency is inversely proportional to the resonating length L of the string  i.e. 

$$f \propto l$$

###### b. The law of the tension (second law): 

The fundamental frequency is proportional to the square root of the stretching force or tension i.e.

$$f \propto T$$

###### c. The law of mass (third law):

The fundamental frequency is inversely proportional to the square root of the mass per unit length i.e.

$$f\propto m$$
