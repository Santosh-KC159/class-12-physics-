#### Mechanical Waves

##### 2.1 Sound wave

Sound is a form of energy which gives sensation of hearing.It travels in the form of mechanical wave so it requires a contineous,elastic and inertial medium.Sound is produced by vibration of sounding body.The sound shows,reflection,refraction,interference and diffraction but not polarization.The polarization of waves confirms its transverse nature and sound is a longitudinal wave so it doesnot show polarization.


##### Reflection of sound 

Like all types of waves,sound wave also exhibits the phenomenon of reflection.It obeys the laws of reflection like light.Echo is an example of reflection of sound which is heard in large hall and in the neighborhood of hills.For reflection of sound wave,reflector must be large size due to large wavelength of sound.The nature of reflected waves depends upon wherher the reflection takes place at the surface of denser medium or at rarer medium.

###### i. Reflection at a denser medium

Sound is longitudinal wave which travels in the form of compressions and rarefactions.When this wave strikes on a rigid surface(denser medium)compressions reflects back as compression and rarefactions as rarefactions.The nature of sound waves does not alter after reflection but phase reverse i.e. a phase change of $180^{0}$ occurs between the incident and reflected wave.This is called reflection without change of type but with change of phase(Change in direction of particles)

###### ii. Reflection at rarer medium

When a sound wave or any lognitudinal wave strikes the surface of rarer medium,compression comes back as rarefaction and rarefactions is reflected back as compression i.e.,the nature of sound wave alter after reflection but phase remains same.This is called reflection with change of type,but without change of phase.

During the reflection of wave,its speed,frequency and wavelength do not change but its phase may or may not change depending on the nature of boundary separating two media.

In the perfect reflection of a wave,its amplitude and hence its intensity remains same.

###### Applications(uses)of reflection of sound

* Sound reflectors
* Ear trumpet
* Physician's stethoscope
* Echo

##### Echo

When sound is produced in front of a distant big obstacle,it is reflected and under suitable conditions,sound is heard again i.e.,the sound is repeated.This repeated sound is called echo.Echo is heard after reflection from a distant rigid obstacle(like cliff,a hill side,wall of building,edge of a forest etc).Echo is heard distinctly if it reaches the ear at least $\frac{1}{10}$th of a second later than the production of original sound.

If $d$ be the distance between the observer and obstacle,and $v$ is speed of sound ,then the time taken $'t'$ by the sound to reach the obstacle and come back is given by,

$t=\frac{total distance travelled}{speed of sound}$               since   $v=\frac{d}{t}$

$t =\frac{d+d}{v}$

$t =\frac{2d}{v}$

putting $t=\frac{1}{10}$ =0.1 sec and $v =340{m}{s}^{-1}$ at ordinary temperature,we have,

$\therefore$ The minimum distance between the observer and the obstacle for the echo to be heard distinctly $(d)=\frac{v.t}{2}$ =$ \frac{340*0.1}{2}$ =$17m$

Thus,for hearing echo distinctly minimum distance between observer and obstacle should be 17m.If distance is less than 17m ,time interval will be short i.e,less than 0.1 sec that no one be able to distinguish the echo from the original sound.

#### Reverberations

If there are number of reflecting surfaces near one another,a series of reflection may takes place.Such sound is called reverberation.Thus,reverberation of a thunder is due to number of reflection of sound from clouds,rocks,cliffs etc.

#### Refraction of sound

As other waves,sound wave obeys the laws of refraction.In fact,the sound gets refracted in much less amount.That is why,we are unable to hear the sound produced in air form under water surface.The refraction of sound explains why sound waves are easier to hear at night than during day time.In the day time ,the upper layers of air are colder than the layers nears the earth.Since sound travels faster in the higher temperature($v$ $\propto \sqrt{T}$),hence it refects in a direction away from the earth.The intensity of sound wave thus diminishes.At the night time,however,the layers of air near the earth are colder than those higher up and sound waves are now refrated towards the earth,with consequent increase in intensity.During refraction of a wave,speed,wavelength,amplitude and intensity change but frequency and phase remains same.

##### Diffraction of sound

The bending or spreading of wave around the corners of a obstacle is called $diffraction$.Diffraction phenomenon occurs only when the size of the obstacle is of the order of wavelength ofthw wave. sound wave,water wave,radio wave,light wave,all show diffraction.Diffraction is a charecteristics of wave.It is due to the diffraction of sound,we can hear a person hidden behind a wall although we can not see him.It is beacause of the fact that wavelenght of sound is large and size of the obstacle or openings(window,door etc.) are comparable to wave length of sound.

During the diffraction,there is no change in spped,wavelength and frequency of wave,since wave remains in the same medium.Diffraction occurs in both tranverse and longitudinal waves.

### 2.2 Speed of Mechanical Wave

Any material medium,through which sound wave travel,must possesses at least two physical properties:elasticity and inertia.This means the wave velcity $v$ depends on the elastic and inertial properties of the medium.When a body is distrubed,the restoring force which are proportional to the displacement are developed in the body to bring it back to the normal position.This can only happen if the body possesses the property of elasticity.Further,the material medium has some inertia.This causes a physical quantity namely density of the medium to come into picture.Thus,it is assumed that,

$v$ $\propto{E}^{a}$

and $v$ $\propto{\rho}^{b}$

Where $E$ is modulus of elasticity of medium and $\rho$ is its density.

$\therefore$ $v$ $\propto{{E}^{a}{\rho}^{b}}$

or, $v$=${k}{E}^{a}{\rho}^{b}$             .....................(i)


Where $k$ is dimensionless constant which can be determined either by an elaborated theory or by experiment which gives $k=1$

Now,

$[E]$ =$[\frac{stress}{strain}]$


but $stress$ =$\frac{force}{area}$ and $strain$ =$\frac{change in dimension}{original dimension}$ =dimensionless.So,

$[E]$=$[\frac{Force}{Area}]$ = $[\frac{{M}{L}{T}^{-2}}{{L}^{2}}]$ = $[{M}{L}^{-1}{T}^{-2}]$

also, $[\rho]$ =$[{M}{L}^{-3}]$

and $[v]$ =$[{L}{T}^{-1}]$

Thus,from $eq^{n}$ (i) we have,

$[{L}{T}^{-1}]$ =$[{{M}{L}^{-1}{T}^{-2}}]^{a}$ $[{{M}{L}^{-3}}]^{b}$

or,$[{L}{T}^{-1}]$ =$[{{M}^{a+b}{L}^{-a-3b}{T}^{-2a}}]$
 
Equating the powers of $M$,$L$ and $T$,we get

$a+b$ =$0$

$-a-3b$ =$1$

and $-2a$ =$-1$

$\therefore$  $a =\frac{1}{2}$  ,$b =-\frac{1}{2}$

Hence,from $eq^{n}$ (1),we get,

$v =k{E}^{\frac{1}{2}}{\rho}^{-\frac{1}{2}}$

or, $v = 1*\sqrt{\frac{E}{\rho}}$


$\therefore$ $v$=$\sqrt{\frac{E}{\rho}}$ = $\sqrt{\frac{modulus of elasticity}{density}}$   .........(ii)

This is a required expression for speed of sound in medium.

The speed of any mechanical wave,transverse or longitudinal,depends on both an inertial property of the medium(to store K.E.) and an elastic property of the medium(to store P.E.).So $eq^{n}$ (ii) can be written as,

$v   =\sqrt{\frac{Elastic property}{Inertial property}}$


###### For a solid rod:
The wave travels along the rod,the relevant modulus of elasticity is Young's modulus $Y$.Thus for solid,

$v  =\sqrt{\frac{Y}{\rho}}$             ..............(iii)

###### For a gas or liquid:
The wave propagates in all direction,relevant modulus od elasticity is the Bulk modulus $B$.

Thus

$v   =\sqrt{\frac{B}{\rho}}$             ...................(iv)


###### Speed of electromagnetic wave:
The speed of electromagnetic wave $(EM)$ is determined with the formula,

$v  =\sqrt{\frac{1}{{\epsilon}{\mu}}}$

Where $\epsilon$ and $\mu$ are the permitivity and permeability of a medium where $EM$ is propagated.

###### Speed of transverse wave through a stretched string:
Speed of transverse wave through a stretched string is determined with the formula,

$v   =\sqrt{\frac{T}{\mu}}$

Where $T$ is the tension of the string and $\mu$ is the mass per unit length of string.















### Speed of Sound in Gaseous Medium


##### Newton's Formula

Newton assumed that compressions and rarefactions occurs very slowly in gas when sound wave propagates.The temperature at the regon of compression and rarefactions change due to the change in pressure in these regions.Since the process is considered very slow,the heat is eaisly transferred from higher temperature region(at compression region) to lower temperature region(at rarefaction region) so that there is no any temperature difference betwen the region of compression and rarefaction i.e.,the temperature of gars during the propagations of sound waves reamains constant i.e.the condition is isothermal process and therefore the Boyle's law hold.

Thus,for a given mass of gas at pressure $P$ and volume $V$ ,we can write

$PV =Constant$

Differentiating,we get

$d(PV)  =0$

or  $PdV +VdP  =0$

$P$ =$-\frac{dP}{\frac{dV}{V}} $   .................(i)

Accoding to definition of Bulk modulus $(B)$ of a gas,we can write,

$B  =\frac{Volume Stress}{Volume strain}$ = $-\frac{change in pressure}{\frac{change in volume}{original volume}}$ = $-\frac{dP}{\frac{dV}{V}}$    ..................(ii)

From(i) and (ii),we get

$\therefore$ $B  =P$         ......................(iii)

From the expression of speed of sound in a medium,we have

$v  =\sqrt{\frac{B}{\rho}}$      ..............................(iv)


From (ii) and (iv),we get

$v  =\sqrt{\frac{P}{\rho}}$         ...............................(v)

This is Newton's for the speed of sound in gas.For air at NTP,we have,

$P =760 mm of Hg$,

$ =760*10^{-3}*13600*9.8$

$=1.013*10^{5} Nm^{-2}$

and $\rho =1.29 kgm^{-3}$

$\therefore$  $v= \sqrt{(\frac{1.013*10^{5}}{1.29})}$ = $280 {m}{s}^{-1}$

But,experimentally observed value of speed of sound in air at NTP is $332 ms^{-1}$.Due to this difference between theoretical and experimental values,it was thought that Newton's formula must have some weakness and thus needed a necessary correction.After a century,Laplace corrected the above relation with slight mdification.

#### Laplace's Correction

Laplace pointed out that gases are poor conductors of heat;and the compressions and rarefactions occur very fast so that there is no compensation of heat at these regions.So,the air in the path of sound wave suffers temperature changes,i.e. the conditions are adiabatic rather than isothermal.Thus,the relation between pressure and volume of the gas through which a sound wave is passing,is given by the adiabatic equation for a gas,i.e.

$PV^{\gamma}  =Constant $

Where $P$ and $V$ be the original pressure and volume of gas respectively and $\gamma =\frac{c_p}{c_v}$ is the specific heat ratio.

Differentiating,we get

$d(PV^{\gamma}) = 0$

or  ${P}{\gamma}{V^{\gamma-{1}}}{dV}+{V}^{\gamma}{dP} =0 $

Dividing by $V^{\gamma-1}$,we get 

$P\gamma dV +VdP =0$


or    ${\gamma} P =-{\frac{dP}{\frac{dV}{V}}} =B$           since $B  =-(\frac{Change in pressure}{Change in Volume})$            ...............................(vi)

The speed of sound in a medium is given by,

$v  =\sqrt{\frac{B}{\rho}}$          ........................(vii)

From (vi) and (vii),we get,

$v =\sqrt{\frac{\gamma P}{\rho}}$         .........................(viii)

$Eq^{n}$ (viii) is laplace's formula for the speed of sound in a gas.Taking $\gamma =1.41$ for air,the speed of sound in air at NTP is calculated as,


$v_0 =\sqrt{\frac{\gamma P}{\rho_0}}$  =$\sqrt{\gamma}$  $\sqrt{\frac{P}{\rho_0}}$

$ =\sqrt{1.14} *280 =331.6ms{-1}$

This values agrees with the experimental value.Hence,Laplace's formula for the speed of sound in gaeous medium is correct formula with correct assumptions.

#### 2.3 Factors Affecting the Speed of Sound in a Gas

###### i.Effect of Temperature:

For a gas of mass $m$ in a volume$V$,

the speed of sound $(v)$ =$\sqrt{\frac{\gamma P}{\rho}}$
                    
$ =\sqrt{\frac{\gamma PV}{\rho V}}$
                       
$ =\sqrt{\frac{\gamma nRT}{m}}$

$ =\sqrt{\frac{\gamma nRT}{nM}}$,  Since $M$=molar mass of gas

$ =\sqrt{\frac{\gamma RT}{M}}$ 

Since $\gamma $ and $M$ are constant for a gas.

$v \propto \sqrt{T}$

Let $v_1$ and $v_2$ be the speed of sound at temperature $T_1$ and$T_2$ respectively in a gas.The relation of speed and temperature is written as,

$\frac{v_1}{v_2} =\sqrt{\frac{T_1}{T_2}}$                .................(i)


##### ii. Effect of Pressure

It is well known that the change in the pressure of a given mass of a gas would produce a corresponding change in volume.From Boyle's law,

$PV  = Constant$

Since, $V =\frac{m}{\rho}$

$\therefore$  $\frac{P.m}{\rho} =Constant $

For a given mass of a gas $\frac{P}{\rho}$ is a constant.

But, $v=\sqrt{\frac{\gamma P}{\rho}}$

For $\frac{P}{\rho}$ to be constant,$v$ should be constant.So,the speed is independent of any change in the atomspheric pressure so long as that change is sufficiently slow and Boyle's law holds good.

To verify this fact experimentally,it has been found that the speed of sound at high altitudes is the same as at sea-level,although the atmospheric pressure at the two place is different.


##### iii.Effect of density of medium:

we consider two different gases of densities $\rho_1$ and $\rho_2$ ,then

$v_1$= $\sqrt{\frac{\gamma P}{\rho_{1}}}$

and $v_2$=$\sqrt{\frac{\gamma P}{\rho_{2}}}$

Dividing,we get,

$\frac{v_{1}}{V_{2}}$=$\sqrt{\frac{\rho_{1}}{\rho_{2}}}$

or ,$v \propto {\frac{1}{\sqrt {\rho}}}$

Thus,speed of sound in gas is inversely proportional to the square root of the density of the medium.

#### iv.Effect of humidity:

The density of water vapour at NTP =$\frac{18}{2200}$=$0.0008\frac{gm}{cc}$ whereas the density of dry air at NTP=$0.001293 \frac{gm}{cc}$.Therefore,water vapour has a density less than the density of dry air. Consequently, the density of moist air is less than than the density of dry air.Since the speed of sound is more in a medium of lesser density,so that,speed of sound is more than the speed of sound in dry air.
therefore,sound is heard upto longer distance in rainy season as compared to summer season.

#### Note

The gases like $O_{2}$,$N_{2}$are heavier than water vapour.If water vapour is mixed to these heavier gases,the density of dry gas is more than that of moist gas,so the speed of sound is more than that of speed of sound in dry gas.

In case if gases lighter than water vapours,the density of dry gas is less than the density of moist gas,so the speed is more in dry gas than in moist gas.So hydrogen becomes moist,its density incresases so the speed of sound in humid hydrogen is less than speed of sound in dry hydrogen.Among all the gases speed of sound is maximum in hydrogen at same.

#### v. Effect of ampliyude and wave length:

The expression for velocity is independent of the amplitude and wavelength of the sound and therefore we can conclude that velocity does not depend upon the amplitude and wave length.But,this is true only for small amplitude.For waves of large amplitude,it has been found that loud sounds travel faster than ordinary sounds in the immediate neighbourhood of the source.In such a wave,pressure is not proportional to density and the volume elasticity increases as the density is increased by compression and decreases as density is reduced by rarefactions.Consquently,a compression wave travels faster and a rarefaction wave slower,Moreever,due to the rapid compression in early stages ,there will probably be a rise in temperature which will cause an increase of velocity.At a distance from the source, the sound velocity comes to noramal value corresponding to that for small amplitudes.

#### vi.Effect of frequency: 

The speed of sound in air is independent on its frequency.But sound waves of different frequencies travel with the same spped in air,although their wavelenths are different.Hence, speed of sound is not affected by change of frequency if medium and temperature remain constant.

###### 1.In a stormy day a boy observes a lightening flash which is followed by a thunder 3s.How would you estimate the distance of the lightening strike from the boy.(Given velocity of sound on that day=332 $ms^{-1}$,velocity of light=$3*10^{8} ms^{-1}$)

###### Solution

Given,
velocity of sound$(v)$=$332 ms^{-1}$

velocity of light$(c)$=$3*10^{8} ms^{-1}$

Time interval of two events,$\Delta t$=3s

As we know that,light travels faster than sound,so

$\Delta t =t_{2}-t_{1}=3s$

$\frac{d}{v}$-$\frac{d}{c}$=3

Where$d$ is the distance of lightening strike from the body.

$d$$(\frac{1}{v}-\frac{1}{c})=3$

$d$$(\frac{1}{332}-\frac{1}{3*10^{8}})=3$

$d$=$\frac{3*332*3*10^{8}}{(3*10^{8}-332)} =996 m$

$\therefore$,the lightening strike from boy is 996 m away.

###### 3. How does the speed of sound in air changes when temperature rises by $1^{0} C$?

###### Solution:

We Know that,

$\frac{v_{\theta}}{v_{0}}=\sqrt{\frac{T_{\theta}}{T_{0}}}$

We have,$T_{\theta}=1^{0} C$

$T_{0}=273$

Now,
$\frac{v_{\theta}}{v_{0}} =\sqrt{\frac{273+\theta}{273}} = (\frac{273+\theta}{273})^{\frac{1}{2}}$

or,$\frac{v_{\theta}}{v_{0}} =(1+\frac{1}{2}.\frac{\theta}{273})$ =$(1+\frac{\theta}{546})$

since $\theta$=$1^{0} C$ ,then

$v_{\theta}=v_{0}(1+\frac{1}{546})$

or,$v_{\theta}-v_{0}=\frac{v_{0}}{546}$

For air,$v_{0} =332 ms^{-1}$.

$\therefore$ $v_{\theta}-v_{0} =\frac{332}{546}=0.61 ms^{-1}$.

Thus, the speed of sound in air increases roughly by $0.61 ms^{-1}$ per degree centigrade rise in temperature.
